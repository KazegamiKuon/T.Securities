```python
import requests
import json
import pandas as pd
import plotly.graph_objects as go
```

```python
df_format = {"Date": lambda t: dt.datetime.strptime(t,"%Y-%m-%d").strftime("%d/%m"),"Close": lambda t: t/1000,"Open": lambda t: t/1000,"High": lambda t: t/1000,"Low": lambda t: t/1000,"Rate Diff": lambda t: round(t*100,1),"Forlume": lambda t: round(t/1000,1)}
from_cols = ["tradingDate","close","open","relDiff","forlume","high","low"]
format_cols = ["Date","Close","Open","Rate Diff","Forlume","High","Low"]
```

```python
codes = ["DXG","SSI","STB"]
fig = go.Figure()
```

```python
api_path = "https://apipubaws.tcbs.com.vn/stock-insight/v1/transinfo/{}/his"
for code in codes:
  response_api = requests.get(api_path.format(code))
  data = json.loads(response_api.text)
  df = pd.DataFrame(data,columns=from_cols)
  df.columns = format_cols
  df.style.format(df_format)
  fig.add_trace(go.Scatter(x=df["Date"],y=df["Rate Diff"],mode="lines+markers",name=code))
```

```python
fig.show()
```
